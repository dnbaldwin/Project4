```python
# Import the modules
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt
%matplotlib inline
```

---

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
#file_path = Path("CTG3.csv")

file_path = Path("data_preprocessed.csv")
#file_path = Path("data_normal_path.csv")
df = pd.read_csv(file_path)

# Review the DataFrame
df.head()
```

```python
inertia = []

k = list(range(1, 6))

for i in k:
    k_model = KMeans(n_clusters=i, random_state=0)
    k_model.fit(df)
    inertia.append(k_model.inertia_)
```

```python
# Define a DataFrame to hold the values for k and the corresponding inertia
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

# Review the DataFrame
df_elbow.head()
```

```python
# Plot the DataFrame
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

```python
# Create and initialise the K-means model instance for 2 clusters
model = KMeans(n_clusters=3, random_state=1)

# Print the model
model
```

```python
model.fit(df)
```

```python
predictions = model.predict(df)
```

```python
# Create a copy of the DataFrame
predictions_df = df.copy()

# Add a column to the DataFrame that contains the customer_ratings information
predictions_df['prediction'] = predictions

# Review the DataFrame
predictions_df.head()
```

```python
# Plot the data points based on the prediction
predictions_df.hvplot.scatter(
    x="LB", 
    y="MSTV", 
    by="NSP"
)
```
